#### Задание 1: Расширение модели

1. **Создайте три новые модели**:
   - **`Publisher`**: представляет издательство, которое публикует книги.
     - Модель должна содержать поля для названия издательства и страны, где оно находится.
     - Связь: одна книга может быть опубликована только одним издательством, поэтому нужно установить связь "один ко многим" между моделями `Publisher` и `Book`. Это означает, что одно издательство может опубликовать несколько книг.
     
   - **`Store`**: представляет книжный магазин.
     - Добавьте поля для названия магазина и города, в котором он находится.
     - Связь: одна книга может продаваться в нескольких магазинах, и один магазин может продавать несколько книг. Для этого нужно использовать связь "многие ко многим" между моделями `Book` и `Store`.

   - **`Review`**: модель для отзывов на книги.
     - Поля: оценка (числовое значение) и текстовый комментарий, а также дата создания отзыва.
     - Связь: одна книга может иметь много отзывов, поэтому создайте связь "один ко многим" между моделями `Book` и `Review`.

2. **Измените существующую модель `Book`**:
   - Добавьте связь с моделью `Publisher`, чтобы указать, какое издательство опубликовало книгу.
   - Добавьте связь с моделью `Store`, чтобы указать, в каких магазинах продаётся книга.

3. **Регистрация моделей в административной панели**:
   - Убедитесь, что все модели зарегистрированы в админке Django, чтобы можно было добавлять и редактировать данные через административный интерфейс.

#### Задание 2: Выполнение сложных запросов

1. **Напишите запрос для нахождения всех книг, опубликованных издательствами из определённой страны**. Например, нужно найти все книги, выпущенные издательствами, находящимися в России. Чтобы это сделать, потребуется запрос, который найдёт все книги, связанные с издательствами, находящимися в заданной стране.

2. **Получите список всех книг, которые продаются в магазине в определённом городе**. Например, найдите все книги, которые продаются в магазинах города Москва. Это запрос с использованием связи "многие ко многим" между книгами и магазинами.

3. **Найдите все книги, которые имеют среднюю оценку выше определённого значения** (например, выше 4.5). Чтобы это сделать, выполните запрос, который будет учитывать отзывы, связанные с книгами, и вычислять среднее значение оценок.

4. **Подсчитайте количество книг, продающихся в каждом магазине**. Выполните агрегатный запрос, который подсчитает количество книг для каждого магазина.

5. **Найдите магазины, где продаются книги, опубликованные после определённой даты** (например, после 2010 года). Затем отсортируйте эти магазины по количеству книг, продающихся в них.

#### Задание 3: Оптимизация запросов

1. **Используйте метод `select_related()` для оптимизации запросов**. Этот метод помогает сократить количество запросов к базе данных, когда требуется получить данные, связанные с одной записью. Примените его для получения данных о книгах и связанных издательствах.

2. **Используйте метод `prefetch_related()` для работы с отношениями "многие ко многим"**. Этот метод загружает связанные данные для нескольких объектов одновременно. Используйте его, чтобы оптимизировать запросы, которые связаны с книжными магазинами и отзывами на книги.

#### Что нужно отправить:

1. Полное описание созданных моделей, включая объяснение связей между ними.
2. Пошаговые инструкции по созданию тестовых данных для издательств, магазинов и книг через административную панель.
3. Описание выполненных запросов, объясняя их логику и результаты.
4. Пояснения к тому, как была выполнена оптимизация запросов с использованием `select_related()` и `prefetch_related()`.
